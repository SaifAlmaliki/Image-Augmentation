```
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
```

```
# load the image
img = load_img('bird.jpg')

# convert to numpy array
img = img_to_array(img)

# expand dimention to one sample
samples = expand_dims(img, 0)
```

**width_shift_range**

```

# create image data augmentation generator(width_shift_range)
data_gen = ImageDataGenerator(width_shift_range=[-200, 200]) 

# prepate iterator
iter = data_gen.flow(samples, batch_size=1)

# generate samples and plot
for i in range(9):

    # define subplot
    pyplot.subplot(330 + 1 + i)

    # generate batch of image
    batch = iter.next()

    # convert to unsigned integers of viewing
    image = batch[0].astype('uint8')

    #plot raw pixel data
    pyplot.imshow(image)

pyplot.show()
```
